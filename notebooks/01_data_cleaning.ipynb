## Libraries

```python
import pandas as pd
import numpy as np
```

## Raw Data

```python
dest = pd.read_csv("../data/raw/destinations.csv")
hotel = pd.read_csv("../data/raw/Hotels.csv")
reviews = pd.read_csv("../data/raw/Review_db.csv")
```

## Checking Missing Values

```python
dest.isnull().sum()
hotel.isnull().sum()
reviews.isnull().sum()
```

## Cleaning Destinations

```python
dest.drop_duplicates(inplace=True)
dest["Weekly Off"].fillna("Open All Days", inplace=True)
```

## Cleaning Hotesl

```python
hotel.drop_duplicates(inplace=True)
hotel.fillna("Unknown", inplace=True)
```

## Cleaning Reviews

```python
reviews.drop_duplicates(inplace=True)
reviews.fillna("Unknown", inplace=True)
```

## Save Clean Data

```python
dest.to_csv("../data/raw/processed/clean_destinations.csv", index=False)
hotel.to_csv("../data/raw/processed/clean_hotels.csv", index=False)
reviews.to_csv("../data/raw/processed/clean_reviews.csv", index=False)
```
