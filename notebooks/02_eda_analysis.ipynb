```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

## Loading Clean Data

```python
dest = pd.read_csv("../data/raw/processed/clean_destinations.csv")
hotel = pd.read_csv("../data/raw/processed/clean_hotels.csv")
```

## Top Destinations

```python
top_dest = dest.sort_values(
    "Google review rating",
    ascending=False
).head(10)

top_dest
```

## Visualizations

```python
plt.barh(top_dest["Name"],
         top_dest["Google review rating"])

plt.gca().invert_yaxis()
plt.show()
```

## Hotel Price Distribution

```python
plt.hist(hotel["Hotel_Price"], bins=20)
plt.show()
```

## Correlation

```python
sns.heatmap(hotel.corr(numeric_only=True),
            annot=True)
plt.show()
```
